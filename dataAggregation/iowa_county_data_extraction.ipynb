```python
dfPopTmp = dfPops

for index, row in dfPopTmp.iterrows():
    y = row['ctyname'].split()
    y = y[:-1]
    t = ""
    for i in y:
      t += i
      t += " "
    t = t.strip()

    dfPopTmp.loc[index, 'ctyname'] = t
```
